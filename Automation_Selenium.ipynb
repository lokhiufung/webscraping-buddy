```python
import time
import random

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.service import Service as ChromeService

from webdriver_manager.chrome import ChromeDriverManager
```

```python
channel_ids = [
    'homefeed_recommed',
    'homefeed.fashion_v3',
    'homefeed.food_v3',
    'homefeed.cosmetics_v3',
    'homefeed.movie_and_tv_v3',
    'homefeed.career_v3',
    'homefeed.love_v3',
    'homefeed.household_product_v3',
    'homefeed.gaming_v3',
    'homefeed.travel_v3',
    'homefeed.fitness_v3',
]
```

```python
hrefs = set()
```

```python
service = ChromeService(ChromeDriverManager().install())

# Setup Chrome options
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

#run in headless mode
# options.add_argument("--headless")

# disable the AutomationControlled feature of Blink rendering engine
options.add_argument('--disable-blink-features=AutomationControlled')

# disable pop-up blocking
options.add_argument('--disable-popup-blocking')

# start the browser window in maximized mode
options.add_argument('--start-maximized')

# disable extensions
options.add_argument('--disable-extensions')

# disable sandbox mode
options.add_argument('--no-sandbox')

# disable shared memory usage
options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(service=service, options=options)


for channel_id in channel_ids:
    
    driver.get(f'https://www.xiaohongshu.com/explore?channel_id={channel_id}')

    # Wait for the modal to appear, adjust the selector as needed
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'login-container')]")))

    # Wait for the close button to become clickable and click it
    close_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//*[@data-v-7c2d5134]"))
    )
    close_button.click()

    # Wait for the posts to load
    WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'feeds-container')]")))


    n_scrolls = 10
    for j in range(0, n_scrolls):

        # Grab all posts
        post_elements = driver.find_elements("xpath", "//div[contains(@class, 'feeds-container')]/section")

        # Process each post
        for post in post_elements:

            try:
                # Find the <a> tag with class "author" inside each post and extract the href attribute
                author_tag = post.find_element("xpath", ".//a[contains(@class, 'author')]")
                href = author_tag.get_attribute("href")
                hrefs.add(href)
            except Exception as e:
                print(f"An error occurred: {e}")        

        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        
        time.sleep(random.uniform(5, 7))
```

```python
driver.close()
```

```python
import pandas as pd

df_profile = []
for i in range(1, 4):
    df_profile.append(pd.read_csv(f'./profiles/profiles-{i}.csv'))
df_profile = pd.concat(df_profile)
df_profile = df_profile.drop_duplicates()
df_profile.head()
```

```python
def is_influencer(n_fans):
    if n_fans in {'1W+', '1K+'}:
        return True
    return False
```

```python
import json

n_fans = []
for user_interaction in df_profile['user_interactions']:
    user_interaction = json.loads(user_interaction)
    n_fans.append(list(user_interaction[1].values())[0].strip())
df_profile['n_fans'] = n_fans
```

```python
df_profile['is_influencer'] = df_profile['n_fans'].apply(is_influencer)
df_profile[df_profile['is_influencer']].shape
df_profile[df_profile['is_influencer']].to_csv('influencer_sample.csv', index=False)
```
